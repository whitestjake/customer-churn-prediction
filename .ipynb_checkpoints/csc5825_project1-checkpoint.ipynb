```python
# seperating features from the target value
X = df.drop("Exited", axis=1)
y = df['Exited']
```

```python
# splitting data into train_val 80% and test %20
X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size=0.2)

# splitting data into train 70% and validation %10 using test size 0.125 because 10/80 = 0.125
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.125)

#checking set sizes
print('Data Split Breakdown')
print('X Train Data:      ', X_train.shape)
print('X Validation Data: ', X_val.shape)
print('X Test Data:       ', X_test.shape)

print('\nData Split Breakdown')
print('y Train Data:      ', y_train.shape)
print('y Validation Data: ', y_val.shape)
print('y Test Data:       ', y_test.shape)
```

```python
# Categorical and numerical features
cat_features = ['Geography', 'Gender']
num_features = [
    'CreditScore',
    'Age',
    'Tenure',
    'Balance',
    'NumOfProducts',
    'HasCrCard',
    'IsActiveMember',
    'EstimatedSalary'
]

#one hot encode categorical features and scale numerical values
numeric_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(drop='first', handle_unknown='ignore')

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, num_features),
        ('cat', categorical_transformer, cat_features)
    ]
)
```

```python
# starting with logistic regression as a baseline
logistic_reg_pipeline = Pipeline(
    steps=[
        ('preprocess', preprocessor),
        ('model', LogisticRegression(max_iter=300))
    ]
)

logistic_reg_pipeline.fit(X_train, y_train)

# validate
y_val_pred = logistic_reg_pipeline.predict(X_val)
y_val_prob = logistic_reg_pipeline.predict_proba(X_val)[:,1]

print('Logistic Regression - Validation set')
print('-----------------------------------------------------')
print(classification_report(y_val, y_val_pred))
print('ROC-AUC:', roc_auc_score(y_val, y_val_prob))
print('-----------------------------------------------------')
```

```python
# Random forest - tuned with validation set
rf_pipeline = Pipeline(
    steps=[
        ("preprocess", preprocessor),
        ("model", RandomForestClassifier(random_state=42))
    ]
)

rf_params = {
    "model__n_estimators": [200, 400],
    "model__max_depth": [None, 6, 10],
    "model__min_samples_split": [2, 5]
}

rf_grid = GridSearchCV(
    rf_pipeline,
    param_grid=rf_params,
    cv=3,
    scoring="roc_auc",
    n_jobs=-1
)

rf_grid.fit(X_train, y_train)

# Validate
y_val_pred_rf = rf_grid.predict(X_val)
y_val_prob_rf = rf_grid.predict_proba(X_val)[:, 1]


print("Random Forest  Validation set")
print("------------------------------------------------------")
print("Best Params:", rf_grid.best_params_)
print(classification_report(y_val, y_val_pred_rf))
print("ROC-AUC:", roc_auc_score(y_val, y_val_prob_rf))
print("------------------------------------------------------")
```

```python
# XGBOOST  tuning with validation

xgb_pipeline = Pipeline(
    steps=[
        ("preprocess", preprocessor),
        ("model", XGBClassifier(
            eval_metric="logloss",
            use_label_encoder=False,
            random_state=42
        ))
    ]
)

xgb_params = {
    "model__n_estimators": [150, 250],
    "model__learning_rate": [0.03, 0.1],
    "model__max_depth": [3, 5, 7]
}

xgb_grid = GridSearchCV(
    xgb_pipeline,
    param_grid=xgb_params,
    cv=3,
    scoring="roc_auc",
    n_jobs=-1
)

xgb_grid.fit(X_train, y_train)

# Validate
y_val_pred_xgb = xgb_grid.predict(X_val)
y_val_prob_xgb = xgb_grid.predict_proba(X_val)[:, 1]

print("XGBOOST  Validation set")
print("--------------------------------------------------")
print("Best Params:", xgb_grid.best_params_)
print(classification_report(y_val, y_val_pred_xgb))
print("ROC-AUC:", roc_auc_score(y_val, y_val_prob_xgb))
print("--------------------------------------------------")
```

```python
# total model evalutions

# function for testing models in a simple manner
def evaluate_model(model, X_test, y_test, name):
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    print(f'{name} - performance on test set')
    print('------------------------------------------------------------------')
    print(classification_report(y_test, y_pred))
    print('ROC-AUC: ', roc_auc_score(y_test, y_prob))

print('Test results of each model')
evaluate_model(logistic_reg_pipeline, X_test, y_test, 'Logistic Regression')
evaluate_model(rf_grid.best_estimator_, X_test, y_test, 'Random Forest')
evaluate_model(xgb_grid.best_estimator_, X_test, y_test, 'XGBoost')
```

```python
# importance of features in XGBoost
encoded_cat_cols = list(
    xgb_grid.best_estimator_
    .named_steps["preprocess"]
    .named_transformers_["cat"]
    .get_feature_names_out(cat_features)
)

all_features = num_features + encoded_cat_cols

importances = xgb_grid.best_estimator_.named_steps["model"].feature_importances_

feature_importance_df = pd.DataFrame({
    "feature": all_features,
    "importance": importances
}).sort_values(by="importance", ascending=False)

print("-------------------------------------------")
print("Top 15 features XGBoost")
print("-------------------------------------------")
print(feature_importance_df.head(15))
```

```python
# ROC Curve for XGBoost
y_test_prob_xgb = xgb_grid.predict_proba(X_test)[:, 1]   # predicted probabilities

fpr, tpr, thresholds = roc_curve(y_test, y_test_prob_xgb)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"XGBoost (AUC = {roc_auc:.3f})")
plt.plot([0, 1], [0, 1], linestyle="--")  # reference line
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - XGBoost")
plt.legend(loc="lower right")
plt.tight_layout()
plt.show()
```

```python
# Sort by importance
indices = np.argsort(importances)[::-1]
sorted_features = feature_importance_df['feature'][indices]
sorted_importances = importances[indices]


# Plot
plt.figure(figsize=(10, 6))
plt.bar(range(len(sorted_importances)), sorted_importances)
plt.xticks(range(len(sorted_features)), sorted_features, rotation=90)
plt.title("XGBoost Feature Importance")
plt.xlabel("Features")
plt.ylabel("Importance Score")
plt.tight_layout()
plt.show()
```
